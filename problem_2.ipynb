```python
import pandas as pd
import numpy as np
import pm4py
```

```python
df = pd.read_csv('https://github.com/hilaler/predict-the-processing-time-of-video-digital-conversion-process/blob/main/problem_2.csv?raw=true', sep=',')
df = pm4py.format_dataframe(df, case_id='Case ID', activity_key='Activity', timestamp_key='Start Timestamp')
log = pm4py.convert_to_event_log(df)
```

```python
df['Start Timestamp'] = pd.to_datetime(df['Start Timestamp'])
df['Complete Timestamp'] = pd.to_datetime(df['Complete Timestamp'])
df['total time'] = df['Complete Timestamp'] - df['Start Timestamp']
df.head()
```

```python
df = df.drop(columns=["Role"])
df.info()
```

```python
start_activities = pm4py.get_start_activities(df)
start_activities
```

```python
end_activities = pm4py.get_end_activities(df)
end_activities
```

```python
filtered_log = pm4py.filter_end_activities(df, ["Pay invoice"])
filtered_log
```

```python
tracefilter_df_pos = pm4py.filter_event_attribute_values(df, "Resource", ["Pedro Alvares"], level="event", retain=True)
tracefilter_df_pos
```

```python
print("Cases that involved Pedro Alvares as the Resources is: ", len(tracefilter_df_pos.value_counts('Case ID')))
```

```python
tracefilter_settle = pm4py.filter_event_attribute_values(df, "Activity", ["Settle dispute with supplier Purchasing Agent"], level="event", retain=True)
tracefilter_settle.head()
```

```python
print("Cases that involved Settle dispute with supplier Purchasing Agent Activity: ", len(tracefilter_settle.value_counts("Case ID")))
```
